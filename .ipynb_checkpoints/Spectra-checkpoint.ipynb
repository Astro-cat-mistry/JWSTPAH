  Working with the<span style="color:#F5793A;">Spectra

```python
import numpy as np
from astropy.io import fits
from astropy.table import Table
from photutils.aperture import aperture_photometry, CircularAnnulus
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib import colormaps
from astropy.visualization import simple_norm

channels = {
    'CH1': 'jw01328-c1006_t014_miri_ch1-shortmediumlong_s3d.fits',
    'CH2': 'jw01328-c1006_t014_miri_ch2-shortmediumlong_s3d.fits',
    'CH3': 'jw01328-c1006_t014_miri_ch3-shortmediumlong_s3d.fits'
}

annuli_radii = [(0.1, 1.4), (1.4, 2.8), (2.8, 4.2), (4.2, 5.6), (5.6, 7), (7, 8.4), (8.4, 9.8)]

pixel_scale_pc = 37.41
pixel_scale_kpc = pixel_scale_pc / 1000

for ch_label, fits_filename in channels.items():
    hdul = fits.open(fits_filename)
    cube_data = hdul[1].data.astype(np.float64)

    wl_start = hdul[1].header['CRVAL3']
    wl_step = hdul[1].header['CDELT3']
    wavelengths = wl_start + np.arange(cube_data.shape[0]) * wl_step

    wcs_header = hdul[1].header
    x_center_pixel = wcs_header['CRPIX1'] - 1
    y_center_pixel = wcs_header['CRPIX2'] - 1

    collapsed_image = np.mean(cube_data, axis=0)

    extent_x_kpc = collapsed_image.shape[1] * pixel_scale_kpc
    extent_y_kpc = collapsed_image.shape[0] * pixel_scale_kpc

    annuli_radii_kpc = [
        (r_in * pixel_scale_kpc, r_out * pixel_scale_kpc)
        for r_in, r_out in annuli_radii
    ]

    cmap = colormaps.get_cmap('hsv')

    plt.figure(figsize=(12, 8))
    norm = simple_norm(collapsed_image, 'sqrt', percent=88)
    plt.imshow(collapsed_image, origin='lower', cmap='binary', norm=norm,
               extent=[0, extent_x_kpc, 0, extent_y_kpc])
    plt.colorbar(label='Infrared Intensity (MJy/sr)')

    for index, (inner_radius, outer_radius) in enumerate(annuli_radii_kpc):
        color = cmap(index / len(annuli_radii_kpc))
        annulus = CircularAnnulus((x_center_pixel * pixel_scale_kpc, y_center_pixel * pixel_scale_kpc),
                                   r_in=inner_radius, r_out=outer_radius)
        annulus.plot(color=color, lw=3, alpha=0.8)

    plt.scatter(x_center_pixel * pixel_scale_kpc, y_center_pixel * pixel_scale_kpc,
                color='red', marker='x', s=100, label='Center (WCS)')

    scale_bar_length_kpc = 200 / 1000
    scale_bar_width_kpc = 0.025

    scale_bar_x_kpc = extent_x_kpc - scale_bar_length_kpc - 0.2
    scale_bar_y_kpc = extent_y_kpc - 0.2

    plt.gca().add_patch(Rectangle((scale_bar_x_kpc, scale_bar_y_kpc),
                                  scale_bar_length_kpc, scale_bar_width_kpc,
                                  color='black'))

    plt.text(scale_bar_x_kpc + scale_bar_length_kpc / 2,
             scale_bar_y_kpc + 0.1,
             '200 pc',
             color='black', ha='center', fontsize=12)

    plt.xlabel('Right Ascension (kpc)')
    plt.ylabel('Declination (kpc)')
    plt.title(f'Annuli Overlaid on Collapsed 2D Image (in kpc) - {ch_label}')
    plt.show()

    spectra = []

    plt.figure(figsize=(12, 8))
    for idx, (inner_radius, outer_radius) in enumerate(annuli_radii):
        annulus_aperture = CircularAnnulus((x_center_pixel, y_center_pixel), r_in=inner_radius, r_out=outer_radius)

        flux = np.zeros(cube_data.shape[0])

        for i in range(cube_data.shape[0]):
            phot_table = aperture_photometry(cube_data[i, :, :], annulus_aperture)
            flux[i] = phot_table['aperture_sum'][0]

        spectra.append(flux)

        flux_conversion_factor = 1000
        wavelength_step = wl_step
        flux_mJy_per_um = flux * flux_conversion_factor / wavelength_step

        color = cmap(idx / len(annuli_radii))
        plt.plot(wavelengths, flux_mJy_per_um, label=f'{ch_label} Annulus {idx+1}', color=color)

        filename = f'{ch_label}_ring{idx + 1}.fits'
        hdr = fits.Header()
        hdr['CRPIX1'] = 1
        hdr['CRVAL1'] = wl_start
        hdr['CDELT1'] = wl_step
        hdr['CTYPE1'] = 'WAVE'
        hdr['CUNIT1'] = 'um'

        primary_hdu = fits.PrimaryHDU(header=hdr)
        data = Table()
        data['WAVELENGTH'] = wavelengths
        data['FLUX'] = flux
        data['FLUX_ERROR'] = np.ones_like(flux) * 0.1
        data_hdu = fits.BinTableHDU(data=data)

        hdu_list = fits.HDUList([primary_hdu, data_hdu])
        hdu_list.writeto(filename, overwrite=True)
        print(f'Saved: {filename}')

    plt.xlabel(r'$\lambda$ (μm)', fontsize=14)
    plt.ylabel(r'$f_\nu$ (mJy / $\lambda$ (μm))', fontsize=14)
    plt.title(f'Extracted Spectra from Annuli ({ch_label})', fontsize=16)
    plt.legend(loc='upper right')
    plt.show()
```

</div>

```python
def create_association_file(output_name, input_files, filename="combine_input.json"):
    """
    Create a JSON association file for JWST's combine_1d step, 
    and add the missing fields: "exptype" and "asn_pool".
    
    Parameters:
    - output_name (str): The base name for the output file.
    - input_files (list): List of input FITS files to combine.
    - filename (str): Name of the JSON file to save (default: "combine_input.json").
    """

    association = {
        "products": [
            {
                "name": output_name,
                "members": [{"expname": file, "exptype": "science"} for file in input_files]
            }
        ],
        "asn_pool": "placeholder_pool"
    }

    with open(filename, "w") as json_file:
        json.dump(association, json_file, indent=4)
    print(f"Association file saved as '{filename}'")

# Set the output name and input files
output_name = "NGC7469"
input_files = [
    "jw01328-c1006_t014_miri_ch1-shortmediumlong_x1d.fits",
    "jw01328-c1006_t014_miri_ch2-shortmediumlong_x1d.fits",
    "jw01328-c1006_t014_miri_ch3-shortmediumlong_x1d.fits"
]

create_association_file(output_name, input_files, filename="NGC7469.json")
```
