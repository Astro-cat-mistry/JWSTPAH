```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt

# Define the file paths
C1_1D = 'jw01328-c1006_t014_miri_ch1-shortmediumlong_x1d.fits'
C2_1D = 'jw01328-c1006_t014_miri_ch2-shortmediumlong_x1d.fits'
C3_1D = 'jw01328-c1006_t014_miri_ch3-shortmediumlong_x1d.fits'
C4_1D = 'jw01328-c1006_t014_miri_ch4-shortmediumlong_x1d.fits'

files_to_stitch = [C1_1D, C2_1D, C3_1D]

# Function to load FITS data
def load_fits(file):
    with fits.open(file) as hdul:
        data = hdul[1].data
        wavelength = data['WAVELENGTH']
        flux = data['FLUX']
    return wavelength, flux

# Load all spectra
spectra = [load_fits(file) for file in files_to_stitch]

# Initialize the stitched spectrum with the first file
stitched_wavelength = spectra[0][0]
stitched_flux = spectra[0][1]

# Loop through each subsequent spectrum to stitch them together
for i in range(1, len(spectra)):
    current_wavelength = spectra[i][0]
    current_flux = spectra[i][1]

    # Find the overlapping region
    overlap_mask = (current_wavelength >= stitched_wavelength.min()) & (current_wavelength <= stitched_wavelength.max())

    if np.any(overlap_mask):
        # Extract the overlapping part
        overlap_wavelength = current_wavelength[overlap_mask]
        overlap_flux_current = current_flux[overlap_mask]

        # Interpolate the stitched flux onto the overlap wavelengths
        overlap_flux_stitched = np.interp(overlap_wavelength, stitched_wavelength, stitched_flux)

        # Calculate the scaling factor
        scale_factor = np.median(overlap_flux_stitched / overlap_flux_current)

        # Apply the scaling factor to the current spectrum
        current_flux *= scale_factor

        # Remove overlapping wavelengths from the current stitched spectrum to avoid duplication
        stitched_wavelength = stitched_wavelength[stitched_wavelength < overlap_wavelength.min()]
        stitched_flux = stitched_flux[:len(stitched_wavelength)]

    # Append the current wavelength and flux to the stitched spectrum
    stitched_wavelength = np.concatenate((stitched_wavelength, current_wavelength))
    stitched_flux = np.concatenate((stitched_flux, current_flux))

# Plot the stitched spectrum
plt.figure(figsize=(10, 6))
plt.plot(stitched_wavelength, stitched_flux, label='Stitched Spectrum', color='purple')
plt.xlabel('Wavelength (microns)')
plt.ylabel('Flux (Jy)')
plt.title('Stitched Spectrum from MIRI Channels')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from astropy.io import ascii
from astropy.table import Table

# Assuming 'stitched_wavelength' and 'stitched_flux' contain the stitched spectrum data
stitched_spectrum = Table([stitched_wavelength, stitched_flux], names=('WAVELENGTH', 'FLUX'))

# Save the table in ECSV format, which is compatible with PAHfit
output_file = 'stitched_spectrum.ecsv'
ascii.write(stitched_spectrum, output_file, format='ecsv', overwrite=True)

print("Stitched spectrum saved to:", output_file)
```

```python
def plot_2d_slice(data_cube, title):
    slice_index = data_cube.shape[2] // 2  # Take the middle slice
    plt.figure(figsize=(10, 5))
    plt.imshow(data_cube[:, :, slice_index], cmap='viridis', aspect='auto')
    plt.colorbar(label='Intensity')
    plt.title(f'Middle Slice of {title} at Spectrum Index {slice_index}')
    plt.xlabel('Spatial Dimension 1')
    plt.ylabel('Spatial Dimension 2')
    plt.show()

# Plot slices for each cube
plot_2d_slice(C1_3D, "C1")
plot_2d_slice(C2_3D, "C2")
plot_2d_slice(C3_3D, "C3")
plot_2d_slice(C4_3D, "C4")
```

```python
from mpl_toolkits.mplot3d import Axes3D

def plot_3d_scatter(data_cube, title):
    slice_index = data_cube.shape[2] // 2  # Middle slice
    X, Y = np.indices(data_cube[:, :, slice_index].shape)
    Z = data_cube[:, :, slice_index]

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(X.flatten(), Y.flatten(), Z.flatten(), c=Z.flatten(), cmap='viridis', marker='o')
    ax.set_xlabel('X Dimension')
    ax.set_ylabel('Y Dimension')
    ax.set_zlabel('Intensity')
    plt.title(f'3D Scatter Plot of {title} at Spectrum Index {slice_index}')
    plt.show()

# Plot 3D scatter plots for each cube
plot_3d_scatter(C1_3D, "C1")
plot_3d_scatter(C2_3D, "C2")
plot_3d_scatter(C3_3D, "C3")
plot_3d_scatter(C4_3D, "C4")
```

```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
from photutils.aperture import CircularAnnulus, aperture_photometry

# Load the 3D data cube
hdul = fits.open(C1_3D)
cube_data = hdul[1].data  # Use HDU 1 as per your checks

# Ensure cube_data is a float array
cube_data = cube_data.astype(np.float64)

# Define wavelength axis
wl_start = hdul[1].header['CRVAL3']
wl_step = hdul[1].header['CDELT3']
wavelengths = wl_start + np.arange(cube_data.shape[0]) * wl_step

# Use the center coordinates you calculated
x_center, y_center = 24, 20  # Adjust to valid pixel coordinates

# Define annular regions: inner_radius, outer_radius (in pixels)
annuli_radii = [(1, 5), (6, 10), (11, 15)]  # Example with smaller annuli

# Check cube dimensions and center validity
print("Cube dimensions:", cube_data.shape)
print("Center coordinates:", (x_center, y_center))

# Store flux for each annulus
spectra = []

# Loop through each annular region
for inner_radius, outer_radius in annuli_radii:
    # Check that the annulus fits within the image
    if (x_center - outer_radius < 0 or x_center + outer_radius >= cube_data.shape[2] or
        y_center - outer_radius < 0 or y_center + outer_radius >= cube_data.shape[1]):
        print(f"Annulus ({inner_radius}, {outer_radius}) is out of bounds.")
        continue
    
    # Define the annular aperture using CircularAnnulus
    annulus_aperture = CircularAnnulus((x_center, y_center), r_in=inner_radius, r_out=outer_radius)

    # Extract the spectrum for this annulus
    flux = np.zeros(cube_data.shape[0])  # To store flux for each wavelength

    for i in range(cube_data.shape[0]):  # Loop through the spectral axis
        phot_table = aperture_photometry(cube_data[i, :, :], annulus_aperture)
        flux[i] = phot_table['aperture_sum'][0]  # Store the sum of flux in this wavelength

    # Check for NaN values in the spectrum
    if np.any(np.isnan(flux)):
        print(f"Spectrum for Annulus ({inner_radius}, {outer_radius}): contains NaN values!")
    
    # Append the spectrum for this ring to the spectra list
    spectra.append(flux)

# Plot the extracted spectra for each annular region
plt.figure(figsize=(10, 6))
for idx, (inner_radius, outer_radius) in enumerate(annuli_radii):
    if idx < len(spectra):  # Only plot valid spectra
        plt.plot(wavelengths, spectra[idx], label=f'Annulus {inner_radius}-{outer_radius} pixels')

plt.xlabel('Wavelength')
plt.ylabel('Flux')
plt.title('Extracted 1D Spectra from Annular Regions')
plt.legend()
plt.grid()
plt.show()

# Optionally save each extracted spectrum to a fits file
# for idx, flux in enumerate(spectra):
#     hdu = fits.PrimaryHDU(flux)
#     hdu.writeto(f'extracted_spectrum_annulus_{annuli_radii[idx][0]}_{annuli_radii[idx][1]}.fits')
```
