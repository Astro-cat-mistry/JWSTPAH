  Plotting the <span style="color:#F5793A;">spectra
</div>

```python
import numpy as np
from astropy.io import fits
from photutils.aperture import aperture_photometry, CircularAnnulus
import matplotlib.pyplot as plt

hdul = fits.open('jw01328-c1006_t014_miri_ch2-shortmediumlong_s3d.fits')
cube_data = hdul[1].data

cube_data = cube_data.astype(np.float64)

wl_start = hdul[1].header['CRVAL3']
wl_step = hdul[1].header['CDELT3']
wavelengths = wl_start + np.arange(cube_data.shape[0]) * wl_step

wcs_header = hdul[1].header
x_center = wcs_header['CRPIX1'] - 1  # 0-indexed
y_center = wcs_header['CRPIX2'] - 1  # 0-indexed

annuli_radii = [(0.1, 1.4), (1.4, 2.8), (2.8, 4.2), (4.2, 5.6), (5.6, 7),
                (7, 8.4), (8.4, 9.8)]

spectra = []

for idx, (inner_radius, outer_radius) in enumerate(annuli_radii):
    if (x_center - outer_radius < 0 or x_center + outer_radius >= cube_data.shape[2] or
        y_center - outer_radius < 0 or y_center + outer_radius >= cube_data.shape[1]):
        print(f"Annulus ({inner_radius}, {outer_radius}) is out of bounds.")
        continue
    
    annulus_aperture = CircularAnnulus((x_center, y_center), r_in=inner_radius, r_out=outer_radius)

    # Extract the spectrum for this annulus
    flux = np.zeros(cube_data.shape[0])

    for i in range(cube_data.shape[0]):
        phot_table = aperture_photometry(cube_data[i, :, :], annulus_aperture)
        flux[i] = phot_table['aperture_sum'][0]

    spectra.append(flux)

flux_conversion_factor = 1000  # 1 MJy = 1000 mJy
wavelength_step = wl_step

# Extracted spectra for all annuli
plt.figure(figsize=(12, 8))
cmap = plt.get_cmap('hsv')

for idx, flux in enumerate(spectra):
    flux_mJy_per_um = flux * flux_conversion_factor / wavelength_step  # Convert for plotting
    color = cmap(idx / len(annuli_radii))
    plt.plot(wavelengths, flux_mJy_per_um, label=f'Annulus {idx+1}', color=color)
    
plt.xlabel(r'$\lambda$ (μm)', fontsize=14)
plt.ylabel(r'$f_\nu$ (mJy / $\lambda$ (μm))', fontsize=14)
plt.title('Extracted Spectra from Annuli', fontsize=16)
plt.legend(loc='upper right')

plt.show()
```
