```python
warnings.filterwarnings("ignore")

def save_ring_as_specmodel(filename, wavelengths, flux):
    n = len(wavelengths)

    spec_table = Table()
    spec_table['WAVELENGTH'] = wavelengths
    spec_table['FLUX'] = flux
    spec_table['FLUX_ERROR'] = np.ones(n) * 0.1
    spec_table['FLUX_VAR_POISSON'] = np.zeros(n)
    spec_table['FLUX_VAR_RNOISE'] = np.zeros(n)
    spec_table['FLUX_VAR_FLAT'] = np.zeros(n)
    spec_table['SURF_BRIGHT'] = np.zeros(n)
    spec_table['SB_ERROR'] = np.zeros(n)
    spec_table['SB_VAR_POISSON'] = np.zeros(n)
    spec_table['SB_VAR_RNOISE'] = np.zeros(n)
    spec_table['SB_VAR_FLAT'] = np.zeros(n)
    spec_table['DQ'] = np.zeros(n, dtype=np.uint32)
    spec_table['BACKGROUND'] = np.zeros(n)
    spec_table['BKGD_ERROR'] = np.zeros(n)
    spec_table['BKGD_VAR_POISSON'] = np.zeros(n)
    spec_table['BKGD_VAR_RNOISE'] = np.zeros(n)
    spec_table['BKGD_VAR_FLAT'] = np.zeros(n)
    spec_table['NPIXELS'] = np.ones(n)

    model = MultiSpecModel()
    model.spec = [{"spec_table": spec_table}]
    model.meta.instrument.name = 'MIRI'
    model.meta.exposure.type = 'MIR_MRS'
    model.meta.exposure.exposure_time = 1.0
    model.meta.bunit_data = 'MJy/sr'

    model.save(filename, overwrite=True)
    print(f"[MultiSpecModel] Saved: {filename}")

# === FITS input files ===
channels = {
    'CH1': hydrocarbons_dir / 'jw01328-c1006_t014_miri_ch1-shortmediumlong_s3d.fits',
    'CH2': hydrocarbons_dir / 'jw01328-c1006_t014_miri_ch2-shortmediumlong_s3d.fits',
    'CH3': hydrocarbons_dir / 'jw01328-c1006_t014_miri_ch3-shortmediumlong_s3d.fits',
}

annuli_radii = [(0.1, 1.4), (1.4, 2.8), (2.8, 4.2), (4.2, 5.6), (5.6, 7), (7, 8.4), (8.4, 9.8)]
pixel_scale_pc = 37.41
pixel_scale_kpc = pixel_scale_pc / 1000

effective_fwhm_arcsec = {'CH1': 0.22, 'CH2': 0.28, 'CH3': 0.66}
target_fwhm = 0.30

def fwhm_to_sigma(fwhm):
    return fwhm / (2.0 * np.sqrt(2.0 * np.log(2)))

for ch_label, fits_filename in channels.items():
    hdul = fits.open(fits_filename)
    cube_data = hdul[1].data.astype(np.float64)

    wl_start = hdul[1].header['CRVAL3']
    wl_step = hdul[1].header['CDELT3']
    wavelengths = wl_start + np.arange(cube_data.shape[0]) * wl_step

    collapsed_image = np.nanmean(cube_data, axis=0)

    current_fwhm = effective_fwhm_arcsec[ch_label]
    if current_fwhm < target_fwhm:
        sigma_pix = np.sqrt(fwhm_to_sigma(target_fwhm)**2 - fwhm_to_sigma(current_fwhm)**2) / 0.11
        kernel = Gaussian2DKernel(sigma_pix)
        collapsed_image = convolve(collapsed_image, kernel, boundary='extend', nan_treatment='interpolate')
        print(f"[{ch_label}] PSF matched to ~{target_fwhm:.2f}\"")

    collapsed_norm = collapsed_image - np.nanmin(collapsed_image)
    mask = np.isfinite(collapsed_norm)
    if not np.any(mask):
        raise ValueError(f"[{ch_label}] All values are NaN after convolution.")

    total = np.nansum(collapsed_norm[mask])
    y_indices, x_indices = np.indices(collapsed_image.shape)
    x_center_pixel = np.nansum(x_indices[mask] * collapsed_norm[mask]) / total
    y_center_pixel = np.nansum(y_indices[mask] * collapsed_norm[mask]) / total

    extent_x_kpc = collapsed_image.shape[1] * pixel_scale_kpc
    extent_y_kpc = collapsed_image.shape[0] * pixel_scale_kpc

    annuli_radii_kpc = [(r_in * pixel_scale_kpc, r_out * pixel_scale_kpc) for r_in, r_out in annuli_radii]

    cmap = colormaps.get_cmap('hsv')
    plt.figure(figsize=(12, 8))
    norm = simple_norm(collapsed_image, 'sqrt', percent=88)
    plt.imshow(collapsed_image, origin='lower', cmap='binary', norm=norm,
               extent=[0, extent_x_kpc, 0, extent_y_kpc])
    plt.colorbar(label='Infrared Intensity (MJy/sr)')

    for index, (inner_radius, outer_radius) in enumerate(annuli_radii_kpc):
        color = cmap(index / len(annuli_radii_kpc))
        annulus = CircularAnnulus((x_center_pixel * pixel_scale_kpc, y_center_pixel * pixel_scale_kpc),
                                   r_in=inner_radius, r_out=outer_radius)
        annulus.plot(color=color, lw=3, alpha=0.8)

    plt.scatter(x_center_pixel * pixel_scale_kpc, y_center_pixel * pixel_scale_kpc,
                color='red', marker='x', s=100, label='Center (auto)')

    scale_bar_length_kpc = 200 / 1000
    scale_bar_width_kpc = 0.025
    scale_bar_x_kpc = extent_x_kpc - scale_bar_length_kpc - 0.2
    scale_bar_y_kpc = extent_y_kpc - 0.2

    plt.gca().add_patch(Rectangle((scale_bar_x_kpc, scale_bar_y_kpc),
                                  scale_bar_length_kpc, scale_bar_width_kpc,
                                  color='black'))

    plt.text(scale_bar_x_kpc + scale_bar_length_kpc / 2,
             scale_bar_y_kpc + 0.1,
             '200 pc',
             color='black', ha='center', fontsize=12)

    plt.xlabel('Right Ascension (kpc)')
    plt.ylabel('Declination (kpc)')
    plt.title(f'Annuli Overlaid on Collapsed 2D Image (in kpc) - {ch_label}')
    plt.show()

    spectra = []
    plt.figure(figsize=(12, 8))
    for idx, (inner_radius, outer_radius) in enumerate(annuli_radii):
        annulus_aperture = CircularAnnulus((x_center_pixel, y_center_pixel), r_in=inner_radius, r_out=outer_radius)
        flux = np.zeros(cube_data.shape[0])
        for i in range(cube_data.shape[0]):
            phot_table = aperture_photometry(cube_data[i, :, :], annulus_aperture)
            flux[i] = phot_table['aperture_sum'][0]
        spectra.append(flux)

        flux_conversion_factor = 1000
        wavelength_step = wl_step
        flux_mJy_per_um = flux * flux_conversion_factor / wavelength_step

        color = cmap(idx / len(annuli_radii))
        plt.plot(wavelengths, flux_mJy_per_um, label=f'{ch_label} Annulus {idx+1}', color=color)

        filename = hydrocarbons_dir / f"{ch_label}_ring{idx + 1}.fits"
        save_ring_as_specmodel(str(filename), wavelengths, flux)

    plt.xlabel(r'$\lambda$ (μm)', fontsize=14)
    plt.ylabel(r'$f_\nu$ (mJy / $\lambda$ (μm))', fontsize=14)
    plt.title(f'Extracted Spectra from Annuli ({ch_label})', fontsize=16)
    plt.legend(loc='upper right')
    plt.show()
```

```python
warnings.filterwarnings("ignore")

def save_ring_as_specmodel(filename, wavelengths, flux):
    n = len(wavelengths)

    # Add units using astropy Columns (Spectrum1D compatibility)
    spec_table = Table()
    spec_table['WAVELENGTH'] = Column(wavelengths, unit=u.um)
    spec_table['FLUX'] = Column(flux, unit=u.MJy / u.sr)
    spec_table['FLUX_ERROR'] = Column(np.ones(n) * 0.1, unit=u.MJy / u.sr)

    # Add remaining columns required by the schema (dummy values)
    spec_table['FLUX_VAR_POISSON'] = np.zeros(n)
    spec_table['FLUX_VAR_RNOISE'] = np.zeros(n)
    spec_table['FLUX_VAR_FLAT'] = np.zeros(n)
    spec_table['SURF_BRIGHT'] = np.zeros(n)
    spec_table['SB_ERROR'] = np.zeros(n)
    spec_table['SB_VAR_POISSON'] = np.zeros(n)
    spec_table['SB_VAR_RNOISE'] = np.zeros(n)
    spec_table['SB_VAR_FLAT'] = np.zeros(n)
    spec_table['DQ'] = np.zeros(n, dtype=np.uint32)
    spec_table['BACKGROUND'] = np.zeros(n)
    spec_table['BKGD_ERROR'] = np.zeros(n)
    spec_table['BKGD_VAR_POISSON'] = np.zeros(n)
    spec_table['BKGD_VAR_RNOISE'] = np.zeros(n)
    spec_table['BKGD_VAR_FLAT'] = np.zeros(n)
    spec_table['NPIXELS'] = np.ones(n)

    # Wrap in a JWST-compatible MultiSpecModel
    model = MultiSpecModel()
    model.spec = [{"spec_table": spec_table}]
    model.meta.instrument.name = 'MIRI'
    model.meta.exposure.type = 'MIR_MRS'
    model.meta.exposure.exposure_time = 1.0
    model.meta.bunit_data = 'MJy/sr'

    model.save(filename, overwrite=True)
    print(f"[MultiSpecModel] Saved: {filename}")

fits_dir = Path('../3_Hydrocarbons')

channels = {
    'CH1': hydrocarbons_dir / 'jw01328-c1006_t014_miri_ch1-shortmediumlong_s3d.fits',
    'CH2': hydrocarbons_dir / 'jw01328-c1006_t014_miri_ch2-shortmediumlong_s3d.fits',
    'CH3': hydrocarbons_dir / 'jw01328-c1006_t014_miri_ch3-shortmediumlong_s3d.fits',
}


pixel_scale_arcsec = 0.11
pixel_scale_pc = 37.41
pixel_scale_kpc = pixel_scale_pc / 1000
effective_fwhm = {'CH1': 0.22, 'CH2': 0.28, 'CH3': 0.66}
target_fwhm = 0.30
sigma_from_fwhm = lambda fwhm: fwhm / (2.0 * np.sqrt(2.0 * np.log(2)))
r0 = 0.9  # mínimo 3xFWHM
step = 1.4
annuli_radii = [(r0 + i * step, r0 + (i+1) * step) for i in range(7)]

# Load all cubes, match PSFs, determine common FOV
cubes = {}
collapsed_images = {}
min_shape = None

for ch, filename in channels.items():
    hdul = fits.open(filename)
    data = hdul[1].data.astype(np.float64)
    wl_start = hdul[1].header['CRVAL3']
    wl_step = hdul[1].header['CDELT3']
    wavelengths = wl_start + np.arange(data.shape[0]) * wl_step

    if effective_fwhm[ch] < target_fwhm:
        sigma_pix = np.sqrt(sigma_from_fwhm(target_fwhm)**2 - sigma_from_fwhm(effective_fwhm[ch])**2) / pixel_scale_arcsec
        data = np.array([convolve(slice, Gaussian2DKernel(sigma_pix), boundary='extend', nan_treatment='interpolate') for slice in data])
        print(f"[{ch}] PSF matched to ~{target_fwhm}\"")

    collapsed = np.nanmean(data, axis=0)
    collapsed_images[ch] = collapsed
    min_shape = collapsed.shape if min_shape is None else (
        min(min_shape[0], collapsed.shape[0]), min(min_shape[1], collapsed.shape[1])
    )
    cubes[ch] = (data, wavelengths)

# Crop to common FOV
for ch in collapsed_images:
    img = collapsed_images[ch]
    y_trim = (img.shape[0] - min_shape[0]) // 2
    x_trim = (img.shape[1] - min_shape[1]) // 2
    collapsed_images[ch] = img[y_trim:y_trim+min_shape[0], x_trim:x_trim+min_shape[1]]
    cubes[ch] = (cubes[ch][0][:, y_trim:y_trim+min_shape[0], x_trim:x_trim+min_shape[1]], cubes[ch][1])

# Compute common luminosity-weighted center
combined_image = sum(img - np.nanmin(img) for img in collapsed_images.values())
mask = np.isfinite(combined_image)
total = np.nansum(combined_image[mask])
y_idx, x_idx = np.indices(combined_image.shape)
x_center = np.nansum(x_idx[mask] * combined_image[mask]) / total
y_center = np.nansum(y_idx[mask] * combined_image[mask]) / total
print(f"[CENTER] x = {x_center:.2f}, y = {y_center:.2f} (common center)")

# Loop over channels to extract and save annuli
from matplotlib import colormaps
cmap = colormaps.get_cmap('hsv')

for ch in channels:
    data, wavelengths = cubes[ch]
    collapsed = collapsed_images[ch]

    extent_x_kpc = collapsed.shape[1] * pixel_scale_kpc
    extent_y_kpc = collapsed.shape[0] * pixel_scale_kpc
    annuli_radii_kpc = [(r_in * pixel_scale_kpc, r_out * pixel_scale_kpc) for r_in, r_out in annuli_radii]

    plt.figure(figsize=(12, 8))
    norm = simple_norm(collapsed, 'sqrt', percent=88)
    plt.imshow(collapsed, origin='lower', cmap='binary', norm=norm,
               extent=[0, extent_x_kpc, 0, extent_y_kpc])
    plt.colorbar(label='MJy/sr')

    for i, (r_in, r_out) in enumerate(annuli_radii_kpc):
        ann = CircularAnnulus((x_center * pixel_scale_kpc, y_center * pixel_scale_kpc), r_in=r_in, r_out=r_out)
        ann.plot(color=cmap(i / len(annuli_radii)), lw=3, alpha=0.8)

    plt.scatter(x_center * pixel_scale_kpc, y_center * pixel_scale_kpc, color='red', s=100, marker='x')
    plt.xlabel('RA (kpc)')
    plt.ylabel('Dec (kpc)')
    plt.title(f'{ch} annuli (FOV common)')
    plt.show()

    plt.figure(figsize=(12, 8))
    for i, (r_in, r_out) in enumerate(annuli_radii):
        ann_ap = CircularAnnulus((x_center, y_center), r_in=r_in, r_out=r_out)
        flux = np.array([aperture_photometry(slice, ann_ap)['aperture_sum'][0] for slice in data])
        flux_mJy_per_um = flux * 1000 / (wavelengths[1] - wavelengths[0])
        plt.plot(wavelengths, flux_mJy_per_um, label=f'{ch} Annulus {i+1}', color=cmap(i / len(annuli_radii)))
        output_filename = hydrocarbons_dir / f"{ch}_ring{i+1}.fits"
        save_ring_as_specmodel(str(output_filename), wavelengths, flux)

    plt.xlabel("λ (μm)")
    plt.ylabel("fν (mJy / μm)")
    plt.title(f"{ch} - Extracted Annuli Spectra")
    plt.legend()
    plt.show()
```

```python
# Directories
repo_dir = Path().resolve().parent
input_dir = repo_dir / "3_Hydrocarbons"
output_json_dir = repo_dir / "1_Polycyclic"
output_fits_dir = repo_dir / "3_Hydrocarbons"

output_json_dir.mkdir(exist_ok=True)
output_fits_dir.mkdir(exist_ok=True)

def create_association_file(output_name, input_files, filename):
    association = {
        "asn_type": "spec",
        "asn_rule": "combine1d",
        "asn_id": f"a{output_name}",
        "program": "12345",
        "target": "NGC7469",
        "asn_pool": "auto-gen",
        "version_id": "1.0",
        "products": [
            {
                "name": output_name,
                "members": [{"expname": file, "exptype": "science"} for file in input_files]
            }
        ]
    }
    with open(filename, "w") as json_file:
        json.dump(association, json_file, indent=4)
    print(f"[JSON] Created: {filename}")

# === MAIN ===
channels = ['CH1', 'CH2', 'CH3']

for ring_number in range(1, 8):
    input_files = [str(input_dir / f"{ch}_ring{ring_number}.fits") for ch in channels]
    output_name = f"Ring{ring_number}"
    
    json_file = output_json_dir / f"{output_name}.json"
    stitched_file = output_fits_dir / f"{output_name}_combined.fits"

    create_association_file(output_name, input_files, json_file)

    print(f"[Combine1dStep] Stitching: {input_files}")
    Combine1dStep.call(
        json_file,
        output_file=stitched_file,
        output_dir=str(output_fits_dir.as_posix()),
        suffix="_stitched"
    )
```
